```python
df3=df[(df1['G1']>10) & (df1['G2']>10) & (df1['G3']>10)]
```

violin plot is used inorder to show how many marks most of the students got

```python
plt.subplot(1,5,1)
plt.title('G1')
sns.violinplot(y='G1',data=df1,color='blue',linewidth=2)
plt.subplot(1,5,3)
plt.title("G2")
sns.violinplot(y='G2',data=df1,color='red',linewidth=2)
plt.subplot(1,5,5)
plt.title("G3")
sns.violinplot(y='G3',data=df1,color='green',linewidth=2)
```
